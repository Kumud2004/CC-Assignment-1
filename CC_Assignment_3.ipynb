```python
#q6
import pandas as pd
employee_data = {
    "Employee_ID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","Diana","Edward"],
    "Department":["HR","IT","IT","Marketing","Sales"],
    "Age":[29,34,41,28,38],
    "Salary":[50000,70000,65000,55000,60000],
    "Years_of_Experience":[4,8,10,3,12],
    "Joining_Date":["2020-03-15","2017-07-19","2013-06-01","2021-02-10","2010-11-25"],
    "Gender":["Female","Male","Male","Female","Male"],
    "Bonus":[5000,7000,6000,4500,5000],
    "Rating":[4.5,4.0,3.8,4.7,3.5]
    }
df_emp = pd.DataFrame(employee_data)
print(df_emp)
shape = df_emp.shape
print("Shape of the DataFrame:",shape)
print(df_emp.describe())
print(df_emp.head(5))
print(df_emp.tail(3))
print("Average Salary:",df_emp["Salary"].mean())
print("Total Bonus Paid:",df_emp["Bonus"].sum())
print("Youngest Employee's Age:",df_emp["Age"].min())
print("Highest Performance Rating:",df_emp["Rating"].max())
df_sorted = df_emp.sort_values(by="Salary",ascending=False)
print(df_sorted)
def categorize_rating(rating):
  if rating >=4.5:
    return "Excellent"
  elif rating >=4.0:
    return "Good"
  else :
    return "Average"
  df_emp["Performance_Category"] = df_emp["Rating"].apply(categorize_rating)
  print(df_emp)
  print(df_emp.isnull().sum())
  df_emp.rename(columns={"Employee_ID":"ID"},inplace=True)
  print(df_emp)
  print(df_emp[df_emp["Years_of_Experience"]>5])
  print(df_emp[df_emp["Department"] =="IT"])
  df_emp["Tax"] = df_emp["Salary"]*0.10
  print(df_emp)
  df_emp.to_csv("Modified_Employee.csv",index=False)
```
