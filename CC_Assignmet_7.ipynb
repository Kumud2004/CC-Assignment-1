```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(102316088)
sales_data = np.random.randint(1000,5001, size=(12, 4))
columns = ['Electronics','Clothing','Home&Kitchen','Sports']
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df = pd.DataFrame(data=sales_data, index=months, columns=columns)
df
```

```python
print(df.head())
print(df.describe())
total_sales_per_category = df.sum()
total_sales_per_month = df.sum(axis=1)
avg_growth_per_category = df.pct_change().mean()
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate (%)'] = df['Total Sales'].pct_change() * 100
df['Electronics(Discounted)'] = df['Electronics'] * 0.90
```

```python
df[columns].plot(figsize=(12,6),marker='o')
plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.legend(title='Category')
plt.show
plt.figure(figsize=(10,6))
sns.boxplot(data=df[columns])
plt.title('Distribution of Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```
