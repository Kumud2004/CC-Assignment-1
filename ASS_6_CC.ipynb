```python
#q1
import numpy as np
import matplotlib.pyplot as plt
M = float(input("Enter a value for M:"))
x = np.linspace(-10,10,400)
y1 = M*x**2
y2 = M*np.sin(x)
plt.figure(figsize=(8,5))
plt.plot(x,y1,label=r"$M\cdot x^2$",color ='r',linestyle='--')
plt.plot(x,y2,label=r"$M\cdot \sin(x)$",color ='b',linestyle='-')
plt.title("Mathematical Function")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#q2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {"Subjects":["Math","Science","English","History","Art"],
        "Scores":[85,90,78,88,92]}
df=pd.DataFrame(data)
plt.figure(figsize=(8,5))
ax = sns.barplot(x="Subjects",y="Scores",data=df,palette="viridis")
for i,score in enumerate(df["Scores"]):
            ax.text(i,score+1,str(score),ha="center",va='bottom')
plt.title("Student's Performance")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis="y",linestyle="--",alpha=0.7)
plt.show()
```

```python
#q3
import numpy as np
import matplotlib.pyplot as plt
roll_number = 102316088
np.random.seed(roll_number)
data = np.random.randn(50)
cumulative_sum =np.cumsum(data)
fig,axes =plt.subplots(2,2,figsize=(10,8))
axes[0,0].plot(cumulative_sum,color='b',marker='o',linestyle='-')
axes[0,0].set_title("Cumulative Sum Plot")
axes[0,0].set_xlabel("Index")
axes[0,0].set_ylabel("Cumulative Sum")
axes[0,0].grid(True)
axes[0,1].scatter(range(50),data,color='r',alpha=0.6)
axes[0,1].set_title("Random Data Scatter Plot")
axes[0,1].set_xlabel("Index")
axes[0,1].set_ylabel("Random Data")
axes[0,1].grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

# 1. Line plot of Total Profit
plt.figure(figsize=(8, 5))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker="o", color="g")
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

# 2. Multi-line plot for product sales
plt.figure(figsize=(10, 6))
for column in df.columns[1:-1]:  # Exclude month_number and total_profit
    sns.lineplot(x=df["month_number"], y=df[column], label=column)

plt.title("Product Sales Over Months")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for all attributes
df.set_index("month_number").plot(kind="bar", figsize=(12, 6), colormap="tab10")
plt.title("Monthly Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.grid(axis="y", linestyle="--")
plt.show()
```
